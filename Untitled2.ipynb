````python
import random
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
````

* (C & E)**5% 0.045 & 10% 0.090** RTP 

* : ` C` : 

    - : `draw_number` `` `draw_number`5

* ` D` (): 

    - : `won_last_draw` False if   `won_last_draw`  True True 
    
*  ` E` : 

    - : `consecutive_losses` ```consecutive_losses` 19
    
* (?):

    -  5% :  10% => 1:2 1/3(33%) 2/3(67%)
    
* :

    -  0  1  `rand_num` `rand_num`****


````python
def simulate_one_draw(draw_number, consecutive_losses, won_last_draw, config):
    """"""
    current_p1 = config['BASE_P1']
    current_p2 = config['BASE_P2']

    #  D: 
    if won_last_draw:
        current_p1 /= 2
        current_p2 /= 2

    #  C & E: 
    is_pity_draw = False
    if draw_number <= 5 and consecutive_losses == 4:
        is_pity_draw = True  # 
    if consecutive_losses >= 19:
        is_pity_draw = True  # 

    # 
    if is_pity_draw:
        return "Prize 1" if random.random() < 1/3 else "Prize 2"
    
    # 
    else:
        rand_num = random.random()
        if rand_num < current_p1:
            return "Prize 1"  # 
        elif rand_num < current_p1 + current_p2:
            return "Prize 2"  # 
        else:
            return "Loss"
````

#### 

````python
def run_full_simulation(prize_2_amount, config):
    """
    
    
    input: 
    output:  7  (Tuple)
          1.  
          2. 95% 
          3. 95% 
          4.  
          5.  
          6. 
          7.  
    """
    prize_1_amount = prize_2_amount * config['PRIZE_RATIO']
    
    winnings_per_draw = np.zeros(config['NUM_SIMULATIONS'])
    
    prize_1_wins = 0
    prize_2_wins = 0
    consecutive_losses = 0
    won_last_draw = False
    
    prize_1_win_indices = [] # ()

#    ==============================================
#                   
#    ==============================================
    for i in range(config['NUM_SIMULATIONS']):
        #  (Sampling)
        result = simulate_one_draw(i + 1, consecutive_losses, won_last_draw, config)
        
        # (Function Evaluation)
        # ()
        if result == "Prize 1":
            winnings_per_draw[i] = prize_1_amount
            consecutive_losses = 0
            won_last_draw = True
            prize_1_wins += 1
            prize_1_win_indices.append(i) # 
        elif result == "Prize 2":
            winnings_per_draw[i] = prize_2_amount
            consecutive_losses = 0
            won_last_draw = True
            prize_2_wins += 1
        else:
            winnings_per_draw[i] = 0
            consecutive_losses += 1
            won_last_draw = False

            
#   ==============================================
#      
#   ==============================================

#   (EV)
    simulated_ev = np.mean(winnings_per_draw)
    
    std_dev = np.std(winnings_per_draw)
    
    #  95% 
    # Z 1.96  95% 
    std_error = std_dev / np.sqrt(config['NUM_SIMULATIONS'])
    ci_lower = simulated_ev - 1.96 * std_error
    ci_upper = simulated_ev + 1.96 * std_error
    
    # 
    overall_p1 = prize_1_wins / config['NUM_SIMULATIONS']
    overall_p2 = prize_2_wins / config['NUM_SIMULATIONS']
    
    # 
    avg_draws_per_prize1 = None
    intervals = []
    if len(prize_1_win_indices) > 1:
        intervals = np.diff(prize_1_win_indices)
        avg_draws_per_prize1 = np.mean(intervals)
    
    return (simulated_ev, ci_lower, ci_upper, overall_p1, overall_p2, avg_draws_per_prize1, intervals)
````

#### 

````python
def plot_results(search_history, final_results, target_ev):
    """"""
    # 
    try:
        font = fm.FontProperties(fname='C:/Windows/Fonts/msjh.ttc', size=14)
    except FileNotFoundError:
        try:
            font = fm.FontProperties(fname='/System/Library/Fonts/PingFang.ttc', size=14)
        except FileNotFoundError:
            print("")
            font = fm.FontProperties(size=14)
    plt.rcParams['axes.unicode_minus'] = False
    
    # EV 
    iterations = [h['iteration'] for h in search_history]
    evs = [h['simulated_ev'] for h in search_history]

    plt.figure(figsize=(12, 6))
    plt.plot(iterations, evs, marker='o', linestyle='-', label='EV')
    plt.axhline(y=target_ev, color='r', linestyle='--', label=f'EV ({target_ev:.2f})')
    plt.title(f'EV  (RTP: {final_results["rtp"]:.0%})', fontproperties=font, fontsize=16)
    plt.xlabel('', fontproperties=font)
    plt.ylabel(' (EV)', fontproperties=font)
    plt.xticks(iterations)
    plt.grid(True)
    plt.legend(prop=font)
    plt.show()

    # 
    low_bounds = [h['low_bound'] for h in search_history]
    high_bounds = [h['high_bound'] for h in search_history]

    plt.figure(figsize=(12, 6))
    plt.plot(iterations, low_bounds, marker='^', linestyle=':', color='blue', label='')
    plt.plot(iterations, high_bounds, marker='v', linestyle=':', color='green', label='')
    plt.fill_between(iterations, low_bounds, high_bounds, color='gray', alpha=0.2, label='')
    plt.title('', fontproperties=font)
    plt.xlabel('', fontproperties=font)
    plt.ylabel(' ()', fontproperties=font)
    plt.xticks(iterations)
    plt.grid(True)
    plt.legend(prop=font)
    plt.show()
    
    #  ()
    intervals = final_results['intervals']
    if len(intervals) > 1:
        plt.figure(figsize=(12, 6))
        #  bin 
        plt.hist(intervals, bins=range(0, int(max(intervals) + 10), 2), edgecolor='black')
        plt.axvline(x=final_results['avg_draws'], color='r', linestyle='--', linewidth=2, label=f": {final_results['avg_draws']:.2f} ")
        plt.title(' ()', fontproperties=font, fontsize=16)
        plt.xlabel('', fontproperties=font)
        plt.ylabel('', fontproperties=font)
        plt.grid(axis='y', alpha=0.75)
        plt.legend(prop=font)
        plt.show()
````

````python
def find_optimal_prizes(target_rtp, config):
    """
    RTP( )
    
    """
    target_ev = config['COST_PER_DRAW'] * target_rtp
    print(f"\n==========================================================")
    print(f" RTP {target_rtp:.0%} (EV={target_ev:.2f}) ...")
    print(f"==========================================================")
    
    low_prize_guess = 1.0
    high_prize_guess = config['COST_PER_DRAW'] * 10 #  Binary search upper bound: 10 
    best_prize_2 = 0
    search_history = []

    for i in range(15):
        mid_prize_guess = (low_prize_guess + high_prize_guess) / 2
        if mid_prize_guess == 0: break
        
        print(f"\n {i+1} :  = {mid_prize_guess:.2f} ")
        
        #  7 
        simulated_ev, _, _, _, _, _, _ = run_full_simulation(mid_prize_guess, config)
        
        search_history.append({
            'iteration': i + 1, 
            'prize_guess': mid_prize_guess, 
            'simulated_ev': simulated_ev,
            'low_bound': low_prize_guess,
            'high_bound': high_prize_guess
        })
        
        print(f" ->  = {simulated_ev:.4f} ")
        
        if simulated_ev > target_ev:
            print(" -> ")
            high_prize_guess = mid_prize_guess
        else:
            print(" -> ")
            low_prize_guess = mid_prize_guess
            
        best_prize_2 = mid_prize_guess

    print("\n\n---  ---")

    #  7 
    final_ev, final_ci_low, final_ci_upper, final_p1, final_p2, final_avg_draws, final_intervals = run_full_simulation(best_prize_2, config)
    final_rtp_val = (final_ev / config['COST_PER_DRAW']) * 100

    print(f"\n (RTP {target_rtp:.0%}):")
    print(f" (X2): {best_prize_2:.2f} ")
    print(f" (X1): {best_prize_2 * config['PRIZE_RATIO']:.2f} ")
    print("--------------------------------------------------")
    print(f" {config['NUM_SIMULATIONS']} ")
    print("\nRTP ")
    print(f" ->  (EV): {final_ev:.4f}  ( {target_ev:.2f} )")
    print(f" -> 95% : [{final_ci_low:.4f}, {final_ci_upper:.4f}] ")
    print(f" ->  (RTP): {final_rtp_val:.2f}% ( {target_rtp:.2%})")
    print("\n")
    print(f" -> : {final_p1:.4%}")
    print(f" -> : {final_p2:.4%}")
    print(f" -> : {(final_p1 + final_p2):.4%}")
    
    print("\n")
    if final_avg_draws:
        print(f" ->  {final_avg_draws:.2f} ")
    else:
        print(" -> ")

    # 
    final_results = {
        'rtp': target_rtp, 
        'best_prize_2': best_prize_2, 
        'avg_draws': final_avg_draws, 
        'intervals': final_intervals
    }
    plot_results(search_history, final_results, target_ev)
````

## main()

````python
# =============================================================================
# /*  (Main Program) */
# =============================================================================
if __name__ == "__main__":
    
    # ---  ---
    # 
    # 5% 0.045 & 10% 0.090 RTP 
    SIMULATION_CONFIG = {
        "COST_PER_DRAW": 50,
        "NUM_SIMULATIONS": 1_000_000, # 
        "BASE_P1": 0.045, 
        "BASE_P2": 0.090, 
        "PRIZE_RATIO": 3.0, #  3 
    }

    # ---  ---
    # RTP
    # [0.90, 0.85, 0.95]
    rtp_targets_to_run = [0.90] 
    
    # RTP
    for rtp in rtp_targets_to_run:
        find_optimal_prizes(rtp, SIMULATION_CONFIG)
````
