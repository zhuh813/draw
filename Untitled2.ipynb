### 規則說明:

* 由於設計要求具有保底機制(規則C & 規則E)，則預設**中頭獎的機率略低於5%為 0.045 & 中貳獎的機率略低於10%為 0.090**，避免 RTP 失衡

* 對於保底機制: `規則 C` 的設計要求: 

    - 設置一個變數: `draw_number`，該變數用於紀錄使用者的 `總抽獎次數`，在每一次抽獎之前，程式會先檢查 `draw_number`在5次之前，會觸發保底

* 對於`規則 D` (避免連續中獎)的設計要求: 

    - 初始一個布林變數: `won_last_draw`為 False，該變數用於紀錄使用者上一次有沒有中獎，並用 if 指令判斷  `won_last_draw` 是否為 True，若為 True ，則把預設中頭獎與貳獎的機率砍半，來避免使用者連續中獎
    
* 對於保底機制 `規則 E` 的設計要求: 

    - 設置一個變數: `consecutive_losses`，該變數用於紀錄使用者的 `連續未抽中的次數`，如果`consecutive_losses` 大於等於19次，則觸發保底
    
* 觸發保底的獎勵機制(要中頭獎還是貳獎?):

    - 以目標頭獎整體的中獎率 5% : 目標貳獎整體的中獎率 10% => 1:2的比例，來決定若觸發保底時，使用者約有 1/3(33%)的機率是拿到頭獎 、2/3(67%)的機率是拿到貳獎
    
* 未觸發保底時的運作機制:

    - 產生一個值域為 0 到 1 之間的隨機數 `rand_num` ，來跟中頭獎的機率與中貳獎的機率比大小，如果`rand_num`皆比中獎機率小，則視為**未中獎**


```python
def simulate_one_draw(draw_number, consecutive_losses, won_last_draw, config):
    """根據當前狀態，模擬一次抽獎並返回結果。"""
    current_p1 = config['BASE_P1']
    current_p2 = config['BASE_P2']

    # 規則 D: 防連續中獎
    if won_last_draw:
        current_p1 /= 2
        current_p2 /= 2

    # 規則 C & E: 檢查是否觸發保底
    is_pity_draw = False
    if draw_number <= 5 and consecutive_losses == 4:
        is_pity_draw = True  # 新手保底
    if consecutive_losses >= 19:
        is_pity_draw = True  # 常規保底

    # 如果觸發保底
    if is_pity_draw:
        return "Prize 1" if random.random() < 1/3 else "Prize 2"
    
    # 如果未觸發保底
    else:
        rand_num = random.random()
        if rand_num < current_p1:
            return "Prize 1"  # 頭獎
        elif rand_num < current_p1 + current_p2:
            return "Prize 2"  # 貳獎
        else:
            return "Loss"
```

#### 蒙地卡羅模擬

```python
def run_full_simulation(prize_2_amount, config):
    """
    執行一次完整的蒙地卡羅模擬，並回報詳細的統計數據。
    
    input: 貳獎的獎金金額。
    output: 一個包含 7 個元素的元組 (Tuple)，依序為：
          1. 模擬期望值 
          2. 95%信賴區間下界 
          3. 95%信賴區間上界 
          4. 頭獎整體中獎率 
          5. 貳獎整體中獎率 
          6. 平均頭獎中獎間隔抽數
          7. 頭獎中獎間隔列表 
    """
    prize_1_amount = prize_2_amount * config['PRIZE_RATIO']
    
    winnings_per_draw = np.zeros(config['NUM_SIMULATIONS'])
    
    prize_1_wins = 0
    prize_2_wins = 0
    consecutive_losses = 0
    won_last_draw = False
    
    prize_1_win_indices = [] # 用於記錄每次抽中頭獎的索引(時刻)

#    ==============================================
#                   蒙地卡羅模擬
#    ==============================================
    for i in range(config['NUM_SIMULATIONS']):
        # 蒙地卡羅步驟一：抽樣 (Sampling)
        result = simulate_one_draw(i + 1, consecutive_losses, won_last_draw, config)
        
        #蒙地卡羅步驟二：函數轉換 (Function Evaluation)
        # 轉換成「數值」(即獎金金額)
        if result == "Prize 1":
            winnings_per_draw[i] = prize_1_amount
            consecutive_losses = 0
            won_last_draw = True
            prize_1_wins += 1
            prize_1_win_indices.append(i) # 記錄中獎時刻
        elif result == "Prize 2":
            winnings_per_draw[i] = prize_2_amount
            consecutive_losses = 0
            won_last_draw = True
            prize_2_wins += 1
        else:
            winnings_per_draw[i] = 0
            consecutive_losses += 1
            won_last_draw = False

            
#   ==============================================
#      根據模擬一百萬次後的結果，計算最終的統計數據
#   ==============================================

#   將一百萬次模擬的獎金結果取平均，得到期望值(EV)的估計量
    simulated_ev = np.mean(winnings_per_draw)
    
    std_dev = np.std(winnings_per_draw)
    
    # 計算 95% 信賴區間
    # Z分數 1.96 對應 95% 的信賴水準
    std_error = std_dev / np.sqrt(config['NUM_SIMULATIONS'])
    ci_lower = simulated_ev - 1.96 * std_error
    ci_upper = simulated_ev + 1.96 * std_error
    
    # 計算整體的頭獎與貳獎中獎率
    overall_p1 = prize_1_wins / config['NUM_SIMULATIONS']
    overall_p2 = prize_2_wins / config['NUM_SIMULATIONS']
    
    # 計算平均中獎間隔
    avg_draws_per_prize1 = None
    intervals = []
    if len(prize_1_win_indices) > 1:
        intervals = np.diff(prize_1_win_indices)
        avg_draws_per_prize1 = np.mean(intervals)
    
    return (simulated_ev, ci_lower, ci_upper, overall_p1, overall_p2, avg_draws_per_prize1, intervals)
```

#### 繪圖

```python
def plot_results(search_history, final_results, target_ev):
    """將搜尋過程與最終結果視覺化。"""
    # 處理中文字體
    try:
        font = fm.FontProperties(fname='C:/Windows/Fonts/msjh.ttc', size=14)
    except FileNotFoundError:
        try:
            font = fm.FontProperties(fname='/System/Library/Fonts/PingFang.ttc', size=14)
        except FileNotFoundError:
            print("警告：未找到指定的中文字體，圖表標題可能顯示為亂碼。")
            font = fm.FontProperties(size=14)
    plt.rcParams['axes.unicode_minus'] = False
    
    # 圖一：EV 收斂過程圖
    iterations = [h['iteration'] for h in search_history]
    evs = [h['simulated_ev'] for h in search_history]

    plt.figure(figsize=(12, 6))
    plt.plot(iterations, evs, marker='o', linestyle='-', label='模擬的EV值')
    plt.axhline(y=target_ev, color='r', linestyle='--', label=f'目標EV ({target_ev:.2f})')
    plt.title(f'EV 收斂過程圖 (目標RTP: {final_results["rtp"]:.0%})', fontproperties=font, fontsize=16)
    plt.xlabel('二分搜尋迭代次數', fontproperties=font)
    plt.ylabel('模擬期望值 (EV)', fontproperties=font)
    plt.xticks(iterations)
    plt.grid(True)
    plt.legend(prop=font)
    plt.show()

    # 圖二：搜尋範圍收縮圖
    low_bounds = [h['low_bound'] for h in search_history]
    high_bounds = [h['high_bound'] for h in search_history]

    plt.figure(figsize=(12, 6))
    plt.plot(iterations, low_bounds, marker='^', linestyle=':', color='blue', label='搜尋下限')
    plt.plot(iterations, high_bounds, marker='v', linestyle=':', color='green', label='搜尋上限')
    plt.fill_between(iterations, low_bounds, high_bounds, color='gray', alpha=0.2, label='有效搜尋範圍')
    plt.title('獎金搜尋範圍收縮圖', fontproperties=font)
    plt.xlabel('二分搜尋迭代次數', fontproperties=font)
    plt.ylabel('貳獎獎金金額 (元)', fontproperties=font)
    plt.xticks(iterations)
    plt.grid(True)
    plt.legend(prop=font)
    plt.show()
    
    # 圖三：頭獎中獎間隔分佈圖 (體感機率)
    intervals = final_results['intervals']
    if len(intervals) > 1:
        plt.figure(figsize=(12, 6))
        # 調整 bin 的範圍以獲得更好的視覺效果
        plt.hist(intervals, bins=range(0, int(max(intervals) + 10), 2), edgecolor='black')
        plt.axvline(x=final_results['avg_draws'], color='r', linestyle='--', linewidth=2, label=f"平均間隔: {final_results['avg_draws']:.2f} 抽")
        plt.title('頭獎中獎間隔分佈圖 (體感機率)', fontproperties=font, fontsize=16)
        plt.xlabel('距離上次中頭獎的抽數', fontproperties=font)
        plt.ylabel('發生次數', fontproperties=font)
        plt.grid(axis='y', alpha=0.75)
        plt.legend(prop=font)
        plt.show()
```

#### 運用蒙地卡羅 + 二分搜尋法

```python
def find_optimal_prizes(target_rtp, config):
    """
    主協調函式：接收一個RTP目標，執行完整的二分搜尋(搜尋 貳獎獎金)與蒙地卡羅模擬，
    最終印出報告並繪製圖表。
    """
    target_ev = config['COST_PER_DRAW'] * target_rtp
    print(f"\n==========================================================")
    print(f"開始為目標 RTP {target_rtp:.0%} (EV={target_ev:.2f}) 進行搜尋...")
    print(f"==========================================================")
    
    low_prize_guess = 1.0
    high_prize_guess = config['COST_PER_DRAW'] * 10 # 設定的 Binary search的 upper bound: 每次抽獎的成本的10倍 
    best_prize_2 = 0
    search_history = []

    for i in range(15):
        mid_prize_guess = (low_prize_guess + high_prize_guess) / 2
        if mid_prize_guess == 0: break
        
        print(f"\n第 {i+1} 次嘗試: 測試貳獎金額 = {mid_prize_guess:.2f} 元")
        
        # 接收全部 7 個回傳值，即使在迴圈中只使用第一個
        simulated_ev, _, _, _, _, _, _ = run_full_simulation(mid_prize_guess, config)
        
        search_history.append({
            'iteration': i + 1, 
            'prize_guess': mid_prize_guess, 
            'simulated_ev': simulated_ev,
            'low_bound': low_prize_guess,
            'high_bound': high_prize_guess
        })
        
        print(f" -> 模擬期望值 = {simulated_ev:.4f} 元")
        
        if simulated_ev > target_ev:
            print(" -> 結果過高，降低獎金範圍。")
            high_prize_guess = mid_prize_guess
        else:
            print(" -> 結果過低，提高獎金範圍。")
            low_prize_guess = mid_prize_guess
            
        best_prize_2 = mid_prize_guess

    print("\n\n--- 搜尋完成，執行最終驗證 ---")

    # 接收全部 7 個回傳值，並賦予有意義的變數名稱
    final_ev, final_ci_low, final_ci_upper, final_p1, final_p2, final_avg_draws, final_intervals = run_full_simulation(best_prize_2, config)
    final_rtp_val = (final_ev / config['COST_PER_DRAW']) * 100

    print(f"\n找到的最優解 (目標RTP {target_rtp:.0%}):")
    print(f"貳獎金額 (X2): {best_prize_2:.2f} 元")
    print(f"頭獎金額 (X1): {best_prize_2 * config['PRIZE_RATIO']:.2f} 元")
    print("--------------------------------------------------")
    print(f"在此設定下，經過 {config['NUM_SIMULATIONS']} 次模擬的最終驗證結果：")
    print("\n【RTP 驗證】")
    print(f" -> 玩家期望值 (EV): {final_ev:.4f} 元 (目標為 {target_ev:.2f} 元)")
    print(f" -> 95% 信賴區間: [{final_ci_low:.4f}, {final_ci_upper:.4f}] 元")
    print(f" -> 玩家回報率 (RTP): {final_rtp_val:.2f}% (目標為 {target_rtp:.2%})")
    print("\n【中獎率驗證】")
    print(f" -> 頭獎整體中獎率: {final_p1:.4%}")
    print(f" -> 貳獎整體中獎率: {final_p2:.4%}")
    print(f" -> 總中獎率: {(final_p1 + final_p2):.4%}")
    
    print("\n【體感機率驗證】")
    if final_avg_draws:
        print(f" -> 玩家平均需要 {final_avg_draws:.2f} 抽，才能獲得一次頭獎。")
    else:
        print(" -> 在本次模擬中，頭獎中獎次數不足以計算平均間隔。")

    # 整理最終結果並呼叫繪圖函式
    final_results = {
        'rtp': target_rtp, 
        'best_prize_2': best_prize_2, 
        'avg_draws': final_avg_draws, 
        'intervals': final_intervals
    }
    plot_results(search_history, final_results, target_ev)
```

## main()

```python
# =============================================================================
# /* 主程式區 (Main Program) */
# =============================================================================
if __name__ == "__main__":
    
    # --- 參數設定值 ---
    # 將所有可調整的參數集中在此，方便管理
    # 由於設計要求具有保底機制，則預設中頭獎的機率略低於5%為 0.045 & 中貳獎的機率略低於10%為 0.090，避免 RTP 失衡
    SIMULATION_CONFIG = {
        "COST_PER_DRAW": 50,
        "NUM_SIMULATIONS": 1_000_000, # 建議使用一百萬次以獲得穩定的「體感機率」數據
        "BASE_P1": 0.045, 
        "BASE_P2": 0.090, 
        "PRIZE_RATIO": 3.0, # 假設頭獎獎金是貳獎獎金的 3 倍
    }

    # --- 跑模擬 ---
    # 在這裡設定您想測試的所有RTP目標
    # 您可以增加或修改這個列表，例如：[0.90, 0.85, 0.95]
    rtp_targets_to_run = [0.90] 
    
    # 迴圈執行，為每一個RTP目標尋找最佳獎金
    for rtp in rtp_targets_to_run:
        find_optimal_prizes(rtp, SIMULATION_CONFIG)
```

#### **三、 商業策略的價值**

正是基於這份高度可靠的數據分析，我們得以將單純的數字，轉化為能夠驅動營收的商業策略。其中，最適合的應用便是結合**行為經濟學**的理論，來設計我們的商品組合。

我們將引入一個經典的理論 **誘餌效應 (Decoy Effect)**

* **核心概念**：當消費者在兩個選項之間猶豫不決時，引入第三個「顯然較差」的選項（誘餌），會讓他們感覺我們真正想賣的那個「目標選項」變得更具吸引力，從而提高其被選擇的機率。

#### **四、 心理導向的禮包設計策略**

根據誘餌效應，並利用玩家唯一能明確感知的**「20抽保底」**規則作為心理錨點，我們設計了以下三種抽獎禮包：

| 禮包名稱 | 內容 (抽數) | 建議定價 | 平均單抽成本 | 戰略角色 |
| :--- | :--- | :--- | :--- | :--- |
| **小資入門包** | **10 抽** | **480 元** | **48 元** | **誘餌 (Decoy)** |
| **核心高手包** | **25 抽** | **1000 元**| **40 元** | **目標 (Target)** |
| **豪華攻頂包** | **50 抽** | **1900 元**| **38 元** | **高階選項 (Upsell)** |

**此設計的心理引導路徑如下：**

1.  **建立價值基準**：「小資入門包」提供了一個基礎的折扣，但其價值感普通。它的存在，是為了讓「核心高手包」顯得更超值。

2.  **凸顯目標選項**：玩家在比較時會發現，「核心高手包」花費約 2 倍的價格，卻能得到 2.5 倍的抽數，其**CP值明顯更高**。更重要的是，25 抽的設計巧妙地**超越了「20抽保底」的週期**，給予玩家一種「購買一個完整的、有保障的遊戲週期」的心理滿足感與確定性。

3.  **完成消費引導**：因此，「小資入門包」成功地扮演了誘餌的角色，在**價格價值**和**目標達成**兩個維度上，凸顯了「核心高手包」的雙重優勢，強力地引導了玩家的消費升級。而「豪華攻頂包」則為高消費玩家提供了追求極致優惠的清晰路徑。

這個策略的根基，依然是我們的模擬數據，因為是這些數據幫助我們確認了整個獎勵系統的穩定性，讓我們可以放心地設計這些心理學上的商業策略。
